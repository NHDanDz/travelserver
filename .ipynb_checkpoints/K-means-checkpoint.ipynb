```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Create a blobs
centers = np.array([
    [-5,-5],
    [-5,5],
    [5,-5],
    [5,5]
])

# Generate 4 blobs based on the centers
blobs, groups = make_blobs(n_samples=200, n_features=2, centers=centers, cluster_std=1.5)
```

```python
# Draw the blobs
plt.scatter(blobs[:, 0], blobs[:, 1], c='gray')
plt.show()
```

# K-means clustering algorithms

```python
# Create random centroid

colors = ['r', 'b', 'g', 'y']
centroids = np.array([
    [-10,1],
    [-3,12],
    [10,10],
    [0,0]
])

# Plot the centroids
for i in range(4):
    plt.scatter(centroids[i:i+1, 0], centroids[i:i+1, 1], c=colors[i], s=250)
plt.scatter(blobs[:, 0], blobs[:, 1], c='gray')
plt.show()
```

```python
for _ in range(200):
    #Step 1: allocate the points based on the centroids
    allocations = []

    for j in range(blobs.shape[0]):
        distances = np.sum((blobs[j] - centroids) ** 2, axis=1)
        argmin = np.argmin(distances)
        allocations.append(argmin)

    #Convert to numpy
    allocations = np.array(allocations)

    # Caculate the centroid
    for j in range(centroids.shape[0]):
        centroids[j] = blobs[allocations == j, :].mean(axis=0)
```

```python
# Create random centroid
colors = ['r', 'b', 'g', 'y']
# Plot the centroids
plt.scatter(blobs[:, 0], blobs[:, 1], c='gray')
for i in range(4):
    plt.scatter(centroids[i:i+1, 0], centroids[i:i+1, 1], c=colors[i], s=250)
plt.show()

```

# Vi k=2

```python
# Import th vin cn thit
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
from matplotlib.patches import Ellipse
```

```python
# Khi to d liu th cng (ta  2D)
data = np.array([
    [1, 1],  # A
    [2, 1],  # B
    [3, 1],  # C
    [4, 1],  # D
    [4, 3],  # E
    [3, 3],  # F
    [2, 3],  # G
    [1, 3]   # H
])
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
```

```python
# Chn s cm K = 2, khi to cm vi A v H lm tm cm
K = 2
centroids = np.array([
    data[0],  # A: (1,1)
    data[7]   # H: (1,3)
])
```

```python
# Hm tnh khong cch v gn im
print("\n--- Vng 1: Gn im ---")
distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
assignments = np.argmin(distances, axis=1)

# Hin th bng khong cch v cm gn
df_round1 = pd.DataFrame(distances, columns=['C1 (A)', 'C2 (H)'])
df_round1['Gn vo cm'] = assignments
print(df_round1)
```

```python
# Cp nht li tm cm theo trung bnh
print("\n--- Vng 1: Cp nht tm cm ---")
new_centroids = np.array([data[assignments == k].mean(axis=0) for k in range(K)])
print(f"Tm cm mi:\nC1: {new_centroids[0]}\nC2: {new_centroids[1]}")
centroids = new_centroids
```

```python
# Gn li im theo tm cm mi
print("\n--- Vng 2: Gn im ---")
distances2 = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
assignments2 = np.argmin(distances2, axis=1)
df_round2 = pd.DataFrame(distances2, columns=['C1 (new)', 'C2 (new)'])
df_round2['Gn vo cm'] = assignments2
print(df_round2)
```

```python
# Cp nht li tm cm ln na
print("\n--- Vng 2: Cp nht tm cm ---")
new_centroids2 = np.array([data[assignments2 == k].mean(axis=0) for k in range(K)])
print(f"Tm cm mi:\nC1: {new_centroids2[0]}\nC2: {new_centroids2[1]}")

# Kim tra hi t
if np.allclose(new_centroids, new_centroids2):
    print("\nThut ton  hi t.\n")
    final_assignments = assignments2
    final_centroids = new_centroids2
else:
    print("\nThut ton cha hi t. Tip tc vng lp...")
```

```python
# Hm v cm v tm cm (km vng khoanh cm m rng)
plt.figure(figsize=(6, 5))
colors = cm.tab10(final_assignments)
plt.scatter(data[:, 0], data[:, 1], c=colors, s=100, label='im')
plt.scatter(final_centroids[:, 0], final_centroids[:, 1], c='red', marker='X', s=200, label='Tm cm')

# Khoanh vng cm vi padding m rng  bao ton b im
for k in range(K):
    cluster_points = data[final_assignments == k]
    min_x, max_x = cluster_points[:, 0].min(), cluster_points[:, 0].max()
    min_y, max_y = cluster_points[:, 1].min(), cluster_points[:, 1].max()
    center_x = (min_x + max_x) / 2
    center_y = (min_y + max_y) / 2
    width = (max_x - min_x) + 1.0
    height = (max_y - min_y) + 1.0
    ellipse = Ellipse((center_x, center_y), width, height,
                      edgecolor='black', fc='none', lw=2, linestyle='--')
    plt.gca().add_patch(ellipse)

# Ghi nhn cc im
for i, txt in enumerate(labels):
    plt.annotate(txt, (data[i, 0]+0.05, data[i, 1]+0.05), fontsize=12)

plt.title('Phn cm K-Means sau khi hi t (c khoanh vng cm y )')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# Vi k = 4

```python
# Import th vin cn thit
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
from matplotlib.patches import Ellipse
```

```python
# Khi to d liu th cng (ta  2D)
data = np.array([
    [1, 1],  # A
    [2, 1],  # B
    [3, 1],  # C
    [4, 1],  # D
    [4, 3],  # E
    [3, 3],  # F
    [2, 3],  # G
    [1, 3]   # H
])
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
```

```python
# Chn s cm K = 4, khi to cm vi A, H, E, D lm tm cm
K = 4
centroids = np.array([
    data[0],  # A: (1,1)
    data[7],  # H: (1,3)
    data[4],  # E: (4,3)
    data[3]   # D: (4,1)
])
```

```python
# Vng lp K-Means n khi hi t
max_iters = 10
for i in range(max_iters):
    print(f"\n--- Vng {i+1}: Gn im ---")
    distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
    assignments = np.argmin(distances, axis=1)
    df = pd.DataFrame(distances, columns=[f'C{k+1}' for k in range(K)])
    df['Gn vo cm'] = assignments
    print(df)

    print(f"\n--- Vng {i+1}: Cp nht tm cm ---")
    new_centroids = np.array([data[assignments == k].mean(axis=0) for k in range(K)])
    for k in range(K):
        print(f"C{k+1}: {new_centroids[k]}")

    # Kim tra hi t
    if np.allclose(centroids, new_centroids):
        print("\nThut ton  hi t.\n")
        break
    centroids = new_centroids
```

```python
# V kt qu sau khi hi t
plt.figure(figsize=(6, 5))
colors = cm.tab10(assignments)
plt.scatter(data[:, 0], data[:, 1], c=colors, s=100, label='im')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Tm cm')

# Khoanh vng cm vi padding m rng  bao ton b im
for k in range(K):
    cluster_points = data[assignments == k]
    if len(cluster_points) > 0:
        min_x, max_x = cluster_points[:, 0].min(), cluster_points[:, 0].max()
        min_y, max_y = cluster_points[:, 1].min(), cluster_points[:, 1].max()
        center_x = (min_x + max_x) / 2
        center_y = (min_y + max_y) / 2
        width = (max_x - min_x) + 1.5
        height = (max_y - min_y) + 1.5
        ellipse = Ellipse((center_x, center_y), width, height,
                          edgecolor='black', fc='none', lw=2, linestyle='--')
        plt.gca().add_patch(ellipse)

# Ghi nhn cc im
for i, txt in enumerate(labels):
    plt.annotate(txt, (data[i, 0]+0.05, data[i, 1]+0.05), fontsize=12)

plt.title('Phn cm K-Means vi K = 4')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
