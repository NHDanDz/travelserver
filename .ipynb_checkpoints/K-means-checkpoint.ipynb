```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Create a blobs
centers = np.array([
    [-5,-5],
    [-5,5],
    [5,-5],
    [5,5]
])

# Generate 4 blobs based on the centers
blobs, groups = make_blobs(n_samples=200, n_features=2, centers=centers, cluster_std=1.5)
```

```python
# Draw the blobs
plt.scatter(blobs[:, 0], blobs[:, 1], c='gray')
plt.show()
```

# K-means clustering algorithms

```python
# Create random centroid

colors = ['r', 'b', 'g', 'y']
centroids = np.array([
    [-10,1],
    [-3,12],
    [10,10],
    [0,0]
])

# Plot the centroids
for i in range(4):
    plt.scatter(centroids[i:i+1, 0], centroids[i:i+1, 1], c=colors[i], s=250)
plt.scatter(blobs[:, 0], blobs[:, 1], c='gray')
plt.show()
```

```python
for _ in range(200):
    #Step 1: allocate the points based on the centroids
    allocations = []

    for j in range(blobs.shape[0]):
        distances = np.sum((blobs[j] - centroids) ** 2, axis=1)
        argmin = np.argmin(distances)
        allocations.append(argmin)

    #Convert to numpy
    allocations = np.array(allocations)

    # Caculate the centroid
    for j in range(centroids.shape[0]):
        centroids[j] = blobs[allocations == j, :].mean(axis=0)
```

```python
# Create random centroid
colors = ['r', 'b', 'g', 'y']
# Plot the centroids
plt.scatter(blobs[:, 0], blobs[:, 1], c='gray')
for i in range(4):
    plt.scatter(centroids[i:i+1, 0], centroids[i:i+1, 1], c=colors[i], s=250)
plt.show()

```

# Với k=2

```python
# Import thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
from matplotlib.patches import Ellipse
```

```python
# Khởi tạo dữ liệu thủ công (tọa độ 2D)
data = np.array([
    [1, 1],  # A
    [2, 1],  # B
    [3, 1],  # C
    [4, 1],  # D
    [4, 3],  # E
    [3, 3],  # F
    [2, 3],  # G
    [1, 3]   # H
])
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
```

```python
# Chọn số cụm K = 2, khởi tạo cụm với A và H làm tâm cụm
K = 2
centroids = np.array([
    data[0],  # A: (1,1)
    data[7]   # H: (1,3)
])
```

```python
# Hàm tính khoảng cách và gán điểm
print("\n--- Vòng 1: Gán điểm ---")
distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
assignments = np.argmin(distances, axis=1)

# Hiển thị bảng khoảng cách và cụm gán
df_round1 = pd.DataFrame(distances, columns=['C1 (A)', 'C2 (H)'])
df_round1['Gán vào cụm'] = assignments
print(df_round1)
```

```python
# Cập nhật lại tâm cụm theo trung bình
print("\n--- Vòng 1: Cập nhật tâm cụm ---")
new_centroids = np.array([data[assignments == k].mean(axis=0) for k in range(K)])
print(f"Tâm cụm mới:\nC1: {new_centroids[0]}\nC2: {new_centroids[1]}")
centroids = new_centroids
```

```python
# Gán lại điểm theo tâm cụm mới
print("\n--- Vòng 2: Gán điểm ---")
distances2 = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
assignments2 = np.argmin(distances2, axis=1)
df_round2 = pd.DataFrame(distances2, columns=['C1 (new)', 'C2 (new)'])
df_round2['Gán vào cụm'] = assignments2
print(df_round2)
```

```python
# Cập nhật lại tâm cụm lần nữa
print("\n--- Vòng 2: Cập nhật tâm cụm ---")
new_centroids2 = np.array([data[assignments2 == k].mean(axis=0) for k in range(K)])
print(f"Tâm cụm mới:\nC1: {new_centroids2[0]}\nC2: {new_centroids2[1]}")

# Kiểm tra hội tụ
if np.allclose(new_centroids, new_centroids2):
    print("\nThuật toán đã hội tụ.\n")
    final_assignments = assignments2
    final_centroids = new_centroids2
else:
    print("\nThuật toán chưa hội tụ. Tiếp tục vòng lặp...")
```

```python
# Hàm vẽ cụm và tâm cụm (kèm vùng khoanh cụm mở rộng)
plt.figure(figsize=(6, 5))
colors = cm.tab10(final_assignments)
plt.scatter(data[:, 0], data[:, 1], c=colors, s=100, label='Điểm')
plt.scatter(final_centroids[:, 0], final_centroids[:, 1], c='red', marker='X', s=200, label='Tâm cụm')

# Khoanh vùng cụm với padding mở rộng để bao toàn bộ điểm
for k in range(K):
    cluster_points = data[final_assignments == k]
    min_x, max_x = cluster_points[:, 0].min(), cluster_points[:, 0].max()
    min_y, max_y = cluster_points[:, 1].min(), cluster_points[:, 1].max()
    center_x = (min_x + max_x) / 2
    center_y = (min_y + max_y) / 2
    width = (max_x - min_x) + 1.0
    height = (max_y - min_y) + 1.0
    ellipse = Ellipse((center_x, center_y), width, height,
                      edgecolor='black', fc='none', lw=2, linestyle='--')
    plt.gca().add_patch(ellipse)

# Ghi nhãn các điểm
for i, txt in enumerate(labels):
    plt.annotate(txt, (data[i, 0]+0.05, data[i, 1]+0.05), fontsize=12)

plt.title('Phân cụm K-Means sau khi hội tụ (có khoanh vùng cụm đầy đủ)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# Với k=3

```python
# Import thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
from matplotlib.patches import Ellipse
```

```python
# Khởi tạo dữ liệu thủ công (tọa độ 2D)
data = np.array([
    [1, 1],  # A
    [2, 1],  # B
    [3, 1],  # C
    [4, 1],  # D
    [4, 3],  # E
    [3, 3],  # F
    [2, 3],  # G
    [1, 3]   # H
])
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
```

```python
# Chọn số cụm K = 3, khởi tạo cụm với A, G, D làm tâm cụm
K = 3
centroids = np.array([
    data[0],  # A: (1,1)
    data[6],  # G: (2,3)
    data[3]   # D: (4,1)
])
```

```python
# Hàm tính khoảng cách và gán điểm
print("\n--- Vòng 1: Gán điểm ---")
distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
assignments = np.argmin(distances, axis=1)

# Hiển thị bảng khoảng cách và cụm gán
df_round1 = pd.DataFrame(distances, columns=['C1 (A)', 'C2 (G)', 'C3 (D)'])
df_round1['Gán vào cụm'] = assignments
print(df_round1)
```

```python
# Cập nhật lại tâm cụm theo trung bình
print("\n--- Vòng 1: Cập nhật tâm cụm ---")
new_centroids = np.array([data[assignments == k].mean(axis=0) for k in range(K)])
for k in range(K):
    print(f"Tâm cụm {k+1}: {new_centroids[k]}")
centroids = new_centroids
```

```python
# Gán lại điểm theo tâm cụm mới
print("\n--- Vòng 2: Gán điểm ---")
distances2 = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
assignments2 = np.argmin(distances2, axis=1)
df_round2 = pd.DataFrame(distances2, columns=['C1 (new)', 'C2 (new)', 'C3 (new)'])
df_round2['Gán vào cụm'] = assignments2
print(df_round2)
```

```python
# Cập nhật lại tâm cụm lần nữa
print("\n--- Vòng 2: Cập nhật tâm cụm ---")
new_centroids2 = np.array([data[assignments2 == k].mean(axis=0) for k in range(K)])
for k in range(K):
    print(f"Tâm cụm {k+1}: {new_centroids2[k]}")

# Kiểm tra hội tụ
if np.allclose(new_centroids, new_centroids2):
    print("\nThuật toán đã hội tụ.\n")
    final_assignments = assignments2
    final_centroids = new_centroids2
else:
    print("\nThuật toán chưa hội tụ. Tiếp tục vòng lặp...")
```

```python
# Hàm vẽ cụm và tâm cụm (kèm vùng khoanh cụm mở rộng)
plt.figure(figsize=(6, 5))
colors = cm.tab10(final_assignments)
plt.scatter(data[:, 0], data[:, 1], c=colors, s=100, label='Điểm')
plt.scatter(final_centroids[:, 0], final_centroids[:, 1], c='red', marker='X', s=200, label='Tâm cụm')

# Khoanh vùng cụm với padding mở rộng để bao toàn bộ điểm
for k in range(K):
    cluster_points = data[final_assignments == k]
    if len(cluster_points) == 0:
        continue
    min_x, max_x = cluster_points[:, 0].min(), cluster_points[:, 0].max()
    min_y, max_y = cluster_points[:, 1].min(), cluster_points[:, 1].max()
    center_x = (min_x + max_x) / 2
    center_y = (min_y + max_y) / 2
    width = (max_x - min_x) + 1.0
    height = (max_y - min_y) + 1.0
    ellipse = Ellipse((center_x, center_y), width, height,
                      edgecolor='black', fc='none', lw=2, linestyle='--')
    plt.gca().add_patch(ellipse)

# Ghi nhãn các điểm
for i, txt in enumerate(labels):
    plt.annotate(txt, (data[i, 0]+0.05, data[i, 1]+0.05), fontsize=12)

plt.title('Phân cụm K-Means với K=3')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# Với k = 4

```python
# Import thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
from matplotlib.patches import Ellipse
```

```python
# Khởi tạo dữ liệu thủ công (tọa độ 2D)
data = np.array([
    [1, 1],  # A
    [2, 1],  # B
    [3, 1],  # C
    [4, 1],  # D
    [4, 3],  # E
    [3, 3],  # F
    [2, 3],  # G
    [1, 3]   # H
])
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
```

```python
# Chọn số cụm K = 4, khởi tạo cụm với A, H, E, D làm tâm cụm
K = 4
centroids = np.array([
    data[0],  # A: (1,1)
    data[7],  # H: (1,3)
    data[4],  # E: (4,3)
    data[3]   # D: (4,1)
])
```

```python
# Vòng lặp K-Means đến khi hội tụ
max_iters = 10
for i in range(max_iters):
    print(f"\n--- Vòng {i+1}: Gán điểm ---")
    distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
    assignments = np.argmin(distances, axis=1)
    df = pd.DataFrame(distances, columns=[f'C{k+1}' for k in range(K)])
    df['Gán vào cụm'] = assignments
    print(df)

    print(f"\n--- Vòng {i+1}: Cập nhật tâm cụm ---")
    new_centroids = np.array([data[assignments == k].mean(axis=0) for k in range(K)])
    for k in range(K):
        print(f"C{k+1}: {new_centroids[k]}")

    # Kiểm tra hội tụ
    if np.allclose(centroids, new_centroids):
        print("\nThuật toán đã hội tụ.\n")
        break
    centroids = new_centroids
```

```python
# Vẽ kết quả sau khi hội tụ
plt.figure(figsize=(6, 5))
colors = cm.tab10(assignments)
plt.scatter(data[:, 0], data[:, 1], c=colors, s=100, label='Điểm')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Tâm cụm')

# Khoanh vùng cụm với padding mở rộng để bao toàn bộ điểm
for k in range(K):
    cluster_points = data[assignments == k]
    if len(cluster_points) > 0:
        min_x, max_x = cluster_points[:, 0].min(), cluster_points[:, 0].max()
        min_y, max_y = cluster_points[:, 1].min(), cluster_points[:, 1].max()
        center_x = (min_x + max_x) / 2
        center_y = (min_y + max_y) / 2
        width = (max_x - min_x) + 1.5
        height = (max_y - min_y) + 1.5
        ellipse = Ellipse((center_x, center_y), width, height,
                          edgecolor='black', fc='none', lw=2, linestyle='--')
        plt.gca().add_patch(ellipse)

# Ghi nhãn các điểm
for i, txt in enumerate(labels):
    plt.annotate(txt, (data[i, 0]+0.05, data[i, 1]+0.05), fontsize=12)

plt.title('Phân cụm K-Means với K = 4')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
