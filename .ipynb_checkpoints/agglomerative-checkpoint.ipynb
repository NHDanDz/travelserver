```python
# Import thư viện
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Khởi tạo dữ liệu 1 chiều
X = np.array([[1], [3], [4], [9], [10], [13], [21], [23], [28], [29]])

# Dùng scipy để tính liên kết và trực quan hóa cây phân cấp
linked = linkage(X, method='single')


# Vẽ dendrogram
plt.figure(figsize=(10, 5))
dendrogram(linked,
           labels=["1", "3", "4", "9", "10", "13", "21", "23", "28", "29"])
plt.title('Agglomerative Clustering Dendrogram (Single Linkage)')
plt.xlabel('Điểm dữ liệu')
plt.ylabel('Khoảng cách hợp nhất')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Cell 6: So sánh các phương pháp liên kết trong Agglomerative trên dữ liệu 1 chiều
X_1d = np.array([[1], [3], [4], [9], [10], [13], [21], [23], [28], [29]])
link_methods = ['single', 'complete', 'average']

for method in link_methods:
    linked = linkage(X_1d, method=method)
    plt.figure(figsize=(10, 4))
    dendrogram(linked,
               labels=[str(x[0]) for x in X_1d],
               distance_sort='ascending')
    plt.title(f'Agglomerative Dendrogram (linkage="{method}")')
    plt.xlabel('Điểm dữ liệu')
    plt.ylabel('Khoảng cách hợp nhất')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
