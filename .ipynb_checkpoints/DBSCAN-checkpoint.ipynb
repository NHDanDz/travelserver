```python
# Import thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
from sklearn.cluster import DBSCAN
```

```python
# Khởi tạo bộ dữ liệu 2D với nhiều điểm hơn
np.random.seed(42)
cluster1 = np.random.normal(loc=[2, 2], scale=0.3, size=(100, 2))
cluster2 = np.random.normal(loc=[6, 2], scale=0.3, size=(100, 2))
cluster3 = np.random.normal(loc=[4, 5], scale=0.3, size=(100, 2))
outliers = np.array([[0, 0], [8, 5], [7, 0], [1, 6], [9, 3]])

# Gộp lại thành một tập dữ liệu
X = np.vstack([cluster1, cluster2, cluster3, outliers])
```

```python
# Tính tay các khoảng cách lân cận để xác định core/border/noise
from scipy.spatial.distance import cdist

print("\n--- Tính tay các khoảng cách Euclidean ---")
eps = 0.6
min_samples = 4
distance_matrix = cdist(X, X)
neighbors_count = np.sum(distance_matrix <= eps, axis=1) - 1  # trừ chính nó

for i, count in enumerate(neighbors_count):
    label = "core" if count >= min_samples else ("border" if count > 0 else "noise")
    print(f"Điểm {i:2d}: Số hàng xóm = {count:2d} → {label}")
```

```python
# Áp dụng DBSCAN bằng sklearn
model = DBSCAN(eps=eps, min_samples=min_samples)
y_pred = model.fit_predict(X)
```

```python
# Trực quan hóa kết quả phân cụm
plt.figure(figsize=(8, 6))
unique_labels = set(y_pred)
colors = [cm.tab10(i) for i in unique_labels if i != -1]
colors.append((0, 0, 0, 1))  # màu đen cho nhiễu

for label in unique_labels:
    cluster_points = X[y_pred == label]
    color = 'black' if label == -1 else cm.tab10(label)
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], c=[color], s=80, label=f'Cụm {label}' if label != -1 else 'Nhiễu')

plt.title('Phân cụm bằng DBSCAN')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Dataset: Iris

```python
# Cell 1: Import thư viện
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
```

```python
# Cell 2: Tải tập dữ liệu IRIS và chọn hai cặp đặc trưng
iris = datasets.load_iris()
X = iris.data
y_true = iris.target

# Cặp đặc trưng 1: petal width và petal length
X_petal = X[:, [2, 3]]

# Cặp đặc trưng 2: sepal width và sepal length
X_sepal = X[:, [0, 1]]

# Chuẩn hóa dữ liệu
scaler = StandardScaler()
X_petal_scaled = scaler.fit_transform(X_petal)
X_sepal_scaled = scaler.fit_transform(X_sepal)
```

```python
# Cell 3: Hàm vẽ kết quả phân cụm
def plot_clusters(X, labels, title):
    plt.figure(figsize=(6, 5))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
    plt.title(title)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Cell 4: KMeans trên dữ liệu petal
kmeans = KMeans(n_clusters=3, random_state=0)
y_kmeans = kmeans.fit_predict(X_petal_scaled)
plot_clusters(X_petal_scaled, y_kmeans, 'KMeans - Petal')
print("Silhouette Score (KMeans - Petal):", silhouette_score(X_petal_scaled, y_kmeans))
```

```python
# Cell 5: Agglomerative trên dữ liệu petal
agg = AgglomerativeClustering(n_clusters=3, linkage='ward')
y_agg = agg.fit_predict(X_petal_scaled)
plot_clusters(X_petal_scaled, y_agg, 'Agglomerative - Petal')
print("Silhouette Score (Agglomerative - Petal):", silhouette_score(X_petal_scaled, y_agg))
```

```python
# Cell 6: DBSCAN trên dữ liệu petal
dbscan = DBSCAN(eps=0.6, min_samples=5)
y_dbscan = dbscan.fit_predict(X_petal_scaled)
plot_clusters(X_petal_scaled, y_dbscan, 'DBSCAN - Petal')
print("Silhouette Score (DBSCAN - Petal):", silhouette_score(X_petal_scaled, y_dbscan))
```

```python
# Cell 7: KMeans trên dữ liệu sepal
kmeans_sepal = KMeans(n_clusters=3, random_state=0)
y_kmeans_sepal = kmeans_sepal.fit_predict(X_sepal_scaled)
plot_clusters(X_sepal_scaled, y_kmeans_sepal, 'KMeans - Sepal')
print("Silhouette Score (KMeans - Sepal):", silhouette_score(X_sepal_scaled, y_kmeans_sepal))
```

```python
# Cell 8: Agglomerative trên dữ liệu sepal
agg_sepal = AgglomerativeClustering(n_clusters=3, linkage='ward')
y_agg_sepal = agg_sepal.fit_predict(X_sepal_scaled)
plot_clusters(X_sepal_scaled, y_agg_sepal, 'Agglomerative - Sepal')
print("Silhouette Score (Agglomerative - Sepal):", silhouette_score(X_sepal_scaled, y_agg_sepal))
```

```python
# Cell 9: DBSCAN trên dữ liệu sepal
dbscan_sepal = DBSCAN(eps=0.6, min_samples=5)
y_dbscan_sepal = dbscan_sepal.fit_predict(X_sepal_scaled)
plot_clusters(X_sepal_scaled, y_dbscan_sepal, 'DBSCAN - Sepal')
print("Silhouette Score (DBSCAN - Sepal):", silhouette_score(X_sepal_scaled, y_dbscan_sepal))
```

```python
# Cell 4: Chạy trên từng cặp đặc trưng
for X_pair, X_pair_indices in zip([X_pair1, X_pair2], [[2, 3], [0, 1]]):
    print(f"\n==== CẶP ĐẶC TRƯNG: {feature_names[X_pair_indices[0]]} vs {feature_names[X_pair_indices[1]]} ====")
    for method in ['kmeans', 'agglomerative', 'dbscan']:
        run_clustering_and_visualize(X_pair, method)
```
